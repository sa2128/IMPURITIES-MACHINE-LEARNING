```python
#import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

#Define the file name
file = "CH4_H2 Finall - Figshare.xlsx"

#Define the sheet
sheet = "70%H2-30%CH4 " # Use the exact sheet name as it appears in the file

#Parameters
viscosity = 'Gas viscosity(cp)'
density = 'Gas Density (Kg/M3)'


#Define the parameter you want to plot
parameter = viscosity

#Define temeparature range
temperature_values = [220,240,260,280,300,320,340,360,380,400]

#import excel sheet into python
df = pd.read_excel(file, sheet_name=sheet)

#Visualize the first five rows
print(df.head())
```

```python
# Data Wrangling
columns = df.columns
print (columns)
```

```python
#drop first three columns
df = df.iloc[:, 3:]
```

```python
#Visaulize available columns
df.head()
columns = df.columns
print (columns)
```

```python
# Define colors enough room has been given to expand temperature range
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'purple']


# Create a figure and axis
fig, ax = plt.subplots()


for temp, color in zip(temperature_values, colors):
    # Filter the DataFrame for the specific temperature
    filtered_df = df[df['Temperature (K)'] == temp]
    
    # Plot viscosity against the defined parameter
    ax.plot(filtered_df['Pressure (Mpa)'], filtered_df[parameter], label=f'Temperature {temp}K', color=color)

# Set the x-label
ax.set_xlabel('Pressure (Mpa)')

#Define the y_label 
ax.set_ylabel(parameter) 
ax.set_title(sheet)
ax.legend()

# Display the plot
plt.grid(True)
plt.show()
```
